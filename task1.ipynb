```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import random


random.seed(24)
num_samples = 15

x = np.array([random.uniform(1, 50) for _ in range(num_samples)]).reshape((-1, 1))
y = np.array([2 * val + random.uniform(-5, 5) for val in x])

plt.scatter(x, y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()


SLR = LinearRegression()
SLR.fit(x, y)
pred = SLR.predict(x)

mse = mean_squared_error(y, pred)
plt.plot(x, pred, color='red', marker='+')

print("MSE:", mse)
print("Intercept:", SLR.intercept_)
print("Slope:", SLR.coef_)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


data = pd.read_excel('/content/drive/MyDrive/data.xlsx')
x = data['x'].values.reshape(-1, 1)
y = data['y'].values


model = LinearRegression()
model.fit(x, y)


slope = model.coef_[0]
intercept = model.intercept_


plt.scatter(x, y, label='Data')
plt.plot(x, model.predict(x), color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression')
plt.show()

print("Slope:", slope)
print("Intercept:", intercept)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


data = pd.read_excel('/content/drive/MyDrive/data (3).xlsx')


X = data.drop('y', axis=1)
y = data['y']


correlations = X.corrwith(y)


model = LinearRegression()
model.fit(X, y)


coefficients = model.coef_
intercept = model.intercept_


print("Coefficients:", coefficients)
print("Intercept:", intercept)


print("Correlation coefficients:")
print(correlations)


y_pred = model.predict(X)


plt.scatter(y, y_pred)
plt.xlabel('Actual y')
plt.ylabel('Predicted y')
plt.title('Actual vs. Predicted')
plt.show()
```
